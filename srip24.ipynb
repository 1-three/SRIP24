```python
import re
import nltk
import spacy
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('punkt')
nlp = spacy.load("en_core_web_sm")
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess(text):
    text = text.lower()

    text = re.sub(r'[^a-zA-Z0-9\s\.]', '', text)

    doc = nlp(text)
    sentences = [sent.text.strip() for sent in doc.sents]

    tokenized_sentences = [nltk.word_tokenize(sentence) for sentence in sentences]

    normalized_sentences = [[token.lower() for token in sentence] for sentence in tokenized_sentences]
    lemmatized_sentences = [[lemmatizer.lemmatize(token) for token in sentence] for sentence in normalized_sentences]
    return lemmatized_sentences
```

```python
def encode(examples):
    if 'text' not in examples:
        raise ValueError("Input examples must contain 'text' key.")

    text = examples['text']

    if isinstance(text, str):
        text = [text]
    elif isinstance(text, list):
        text = [str(t) for t in text]
    else:
        raise ValueError("Input 'text' must be a string or a list of strings.")

    return tokenizer(text, truncation=True, padding='max_length', return_tensors='pt')

encoded_dataset = preprocessed_dataset.map(encode, batched=True)
```
